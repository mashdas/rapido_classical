```
##Note:This approach won't always work...because the diff between lat and lng won't always be so pronounced
##To make it easier i am going to combine lat and lng as a single metric...so that it becomes easier to sample true negatives

def one_metric(x,y):
  return np.sqrt(x**2+y**2)
```

```
X=X[X["number"]==15564.0]
```

```
X.shape
```

```
X["loc"]=X.apply(lambda row:one_metric(row["pick_lat"],row["pick_lng"]),axis=1)
```

```
#3d plotting(lat)

from mpl_toolkits.mplot3d import Axes3D  

import matplotlib.pyplot as plt
import numpy as np


fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

hist, xedges, yedges = np.histogram2d(lat, time, bins=500, )

# Construct arrays for the anchor positions of the 16 bars.
xpos, ypos = np.meshgrid(xedges[:-1] + 0.25, yedges[:-1] + 0.25, indexing="ij")
xpos = xpos.ravel()
ypos = ypos.ravel()
zpos = 0

# Construct arrays with the dimensions for the 16 bars.
dx = dy = 0.5 * np.ones_like(zpos)
dz = hist.ravel()

ax.bar3d(xpos, ypos, zpos, dx, dy, dz, zsort='average')

plt.show()
```
